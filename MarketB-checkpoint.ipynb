```python
#only required value is needed no need of any whitespaces.

cust_level = df[["Member_number","itemDescription"]].sort_values(by="Member_number",ascending=False)
cust_level["itemDescription"]=cust_level["itemDescription"].str.strip()
cust_level
```

```python



transactions=[a[1]['itemDescription'].tolist() for a in list(cust_level.groupby(["Member_number"]))]
```

```python
#importing apyori package for association rules
from apyori import apriori
rules=apriori(transactions = transactions, min_support = 0.002, min_confidence = 0.05, min_lift = 3, min_length=2)
```

```python
results=list(rules)
```

```python
#In Python, frozenset is the same as set except the frozensets are immutable which means that elements from the frozenset 
#cannot be added or removed once created.
#This function takes input as any iterable object and converts them into an immutable object. 
results
```

```python
#creating user defined functions for arranging the itemsets
def inspect(results):
    lhs= [tuple(result[2][0][0])[0] for result in results]
    rhs= [tuple(result[2][0][1])[0] for  result in results]
    supports =[results[1] for result in results]
    confidences =[result[2][0][2] for result in results]
    lifts=[result[2][0][3] for result in results]
    return list(zip(lhs,rhs,supports,confidences,lifts))
resultsindataframe = pd.DataFrame(inspect(results),columns=['Left Hand Size','Right Hand Size','Support','Confidences','Lift'])
```

```python
resultsindataframe.nlargest(n=5, columns="Lift")
```
